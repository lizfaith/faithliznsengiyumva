```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("darkgrid")
```

```python
## Load the dataset
music = pd.read_csv("Spotify_Song_Attributes.csv")
music.head()
```

```python
df = music.drop_duplicates()
df.head()
```

```python
# Correlation matrix
corr_matrix = df.corr(numeric_only = True)
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')
plt.title('Correlation Matrix')
plt.show()
```

```python
## Data Analysis for the most popular artists in 2023.
top_five_artists = df.groupby("artistName").count().sort_values(by="trackName", ascending=False)["trackName"][:5]
top_five_artists
```

```python
top_five_artists.plot.barh(color = "darkcyan")
plt.show()
```

```python
## Top five loudest Tracks
top_five_loudest_tracks = df[["loudness", "trackName"]].sort_values(by="loudness", ascending=True)[:5]
top_five_loudest_tracks
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x="loudness", y="trackName", data=top_five_loudest_tracks, palette="crest")
plt.title("Top 5 loudest tracks")
plt.show()
```

```python
## Artist with the song that people dance to the most.
top_five_artists_danceable_songs = df[["duration_ms", "trackName", "artistName"]].sort_values(by="duration_ms", 
                                                                                            ascending=False)[:5]
top_five_artists_danceable_songs
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x="danceability", y="artistName", data=top_five_artists_danceable_songs, palette = 'GnBu')
plt.title("Artist With The Most Danceable Song")
plt.show()
```

```python
## Most instrumental Tracks
top_ten_instrumental_tracks = df[["instrumentalness", "artistName", "trackName"]].sort_values(by="instrumentalness", ascending=False)[:5]
top_ten_instrumental_tracks
```

```python
plt.figure(figsize=(12, 7))
plt.pie(x="instrumentalness", data=top_ten_instrumental_tracks, autopct='%1.2f%%', labels=top_ten_instrumental_tracks.trackName, colors = sns.color_palette("PuBuGn"))
plt.show()
```

```python
# Duration distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='duration_ms', bins=30, kde=True)
plt.title('Distribution of Track Duration')
plt.xlabel('Duration (ms)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sum the msPlayed for each genre
genre_ms_played = df.groupby('genre')['msPlayed'].sum().reset_index()
genre_ms_played
```

```python
# Find the genre with the highest total msPlayed
most_listened_genre = genre_ms_played.loc[genre_ms_played['msPlayed'].idxmax()]
most_listened_genre
```

```python
# Danceability vs. Energy scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='danceability', y='energy', hue='valence', palette='viridis', alpha=0.6)
plt.title('Danceability vs. Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.legend(title='Valence', loc='upper right', bbox_to_anchor=(1.25, 1))
plt.show()
```
